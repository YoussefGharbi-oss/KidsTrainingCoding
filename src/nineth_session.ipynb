```python
with open("data.txt" , "r") as origin_file :
  contenu_one = origin_file.read()
with open("data_copy.txt" , "r") as my_file :
  contenu_two = my_file.read()

with open("merged.txt" , "w") as merged_file :
  merged_file.write(contenu_one + contenu_two)
```
